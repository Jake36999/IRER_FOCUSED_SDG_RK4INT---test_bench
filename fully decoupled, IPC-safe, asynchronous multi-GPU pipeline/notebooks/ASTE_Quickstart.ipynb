# ASTE Quickstart: Loading and Visualizing Simulation Data
This notebook demonstrates how to load a `rho_history.h5` file, extract the `psi` field, and plot 3D Quantule topological structures using matplotlib.

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Path to your simulation output
rho_history_path = '../simulation_data/rho_history_high_sse.h5'
with h5py.File(rho_history_path, 'r') as f:
    # Extract the final psi field (assumes psi_history is [t, x, y, z])
    psi_history = f['psi_history'][:]
    psi_final = psi_history[-1]  # Last time step
    print('psi_final shape:', psi_final.shape)
```

## Visualize the 3D Quantule Topological Structures
We'll plot the magnitude of the psi field as a 3D scatter plot.

```python
# Threshold for visualization (e.g., top 1% by magnitude)
threshold = np.percentile(np.abs(psi_final), 99)
coords = np.argwhere(np.abs(psi_final) >= threshold)
values = np.abs(psi_final)[np.abs(psi_final) >= threshold]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(coords[:,0], coords[:,1], coords[:,2], c=values, cmap='viridis', s=10)
plt.colorbar(sc, label='|psi|')
ax.set_title('3D Quantule Topological Structures (|psi| > 99th percentile)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
